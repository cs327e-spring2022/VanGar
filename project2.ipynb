```python
# inner join on film, category, and film_category, excluding horror movies
```

```python
!psql -c "select title, name from pagila.film join pagila.film_category on film.film_id=film_category.film_id join pagila.category on film_category.category_id=category.category_id where category.name != 'Horror' order by title limit 10"
```

```python
# inner join on movie titles and language by language_id, descending order by title and including only films released in 2006
```

```python
!psql -c "select title, name from pagila.film join pagila.language on film.language_id=language.language_id where film.release_year = '2006' order by film.title desc limit 10"
```

```python
# inner joining the city and country tables on country_id, ordered by country name descending and excluding Yemen
```

```python
!psql -c "select city, country from pagila.city join pagila.country on city.country_id = country.country_id where country.country != 'Yemen' order by country.country desc limit 10"
```

```python
# pairs of films with same release year, self-join
```

```python
!psql -c "select f1.title, f2.title, f2.release_year from pagila.film f1 join pagila.film f2 on f1.film_id != f2.film_id and f1.release_year = f2.release_year limit 10"
```

```python
# outer joining the film and language tables on language_id, only PG rated movies in English 
```

```python
!psql -c "select title, rating, name from pagila.film full join pagila.language on film.language_id = language.language_id where name = 'English' and rating = 'PG' limit 10"
```

```python
# outer join of language and film tables on language_id, excluding French, ordered by language name, limit 10
```

```python
!psql -c "select name, title from pagila.film full join pagila.language on film.language_id=language.language_id where name != 'French' order by name limit 10"
```

```python
# outer join of customer and payment tables on customer_id, where amount is greater than $5 ordered by amount descending, limit 10
```

```python
!psql -c "select first_name, amount from pagila.customer full join pagila.payment on customer.customer_id = payment.customer_id where amount > '4.99' order by amount desc limit 10"
```
